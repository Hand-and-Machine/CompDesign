```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Let's use matplotlib to plot some points in rectangular coordinates!
## You should already know how rectangular coordinates work - this is just to show how the syntax works.

## The x-coordinates of the points
X = np.array([1, 2, 3, -1, -2, -3])

## The y-coordinates of the points
Y = np.array([-5, -3, -1, 1, 3, 0])

## Setting up the window
fig = plt.figure()
ax = fig.add_subplot(111, projection="rectilinear")
plt.xlim([-7, 7])
plt.ylim([-7, 7])

## Draws the scatter plot
c = ax.scatter(X, Y, alpha = 0.75)
```

```python
## Now let's try this in polar coordinates,
## where each point is specified by its distance from the origin (r) and angle (theta) from the positive x-axis
## rather than its x-position and y-position.

## The r-values of the points
R = np.array([1, 1, 2, 2, 3, 3])
## The theta-values of the points
## Note: these are measured in radians
THETA = np.array([np.pi/2, np.pi, np.pi/3, np.pi, np.pi/4, np.pi])

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')
## Note: the below places an upper limit on r, not y, since r is the independent variable in polar coordinates
plt.ylim([0, 4])

c = ax.scatter(THETA, R, alpha=0.75)
```
